## **Question 2**


Summarize the sales year wise and draw the bar plot.


```python
#- Summarize the sales year wise and draw the bar plot.
import numpy as np
df=pd.read_csv("/data.csv")
table=pd.pivot_table(data=df,index='Year',values='Global_Sales',aggfunc=np.sum)
print(table)
plt.bar(table.index,table['Global_Sales']) 
plt.show()
#19BCN7003
```

Customize the bar plot with color, title, and axis labels.

```python
# Customize the bar plot with color, title, and axis labels.
plt.bar(table.index,table['Global_Sales'],color="red") 
plt.xlabel('Year') 
plt.ylabel('Global Sales') 
plt.title('Global Sales by year') 
plt.show()
#19BCN7003
```

Summarize the sales based on platform and draw the bar 
plot.

```python
# Summarize the sales based on platform and draw the bar plot.
table1=pd.pivot_table(data=df,index='Platform',values='Global_Sales',aggfunc=np.sum)
print(table1)
#bar plot
plt.bar(table1.index,table1['Global_Sales'],color="red") 
plt.xticks(rotation=90)
plt.xlabel('Platform') 
plt.ylabel('Game Global Sales') 
plt.title('Global Sales by Platform') 
plt.show()
#19BCN7003
```

Draw a bar plot for descending order of summarization for 
the sales based on platform.

```python
# Draw a bar plot for descending order of summarization for the sales based on platform.
table1=table1.sort_values('Global_Sales',ascending=False)
plt.bar(table1.index,table1['Global_Sales'],data=table1,color="red") 
plt.xticks(rotation=90)
plt.xlabel('Platform') 
plt.ylabel('Game Global Sales') 
plt.title('Global Sales by Platform') 
plt.show()
#19BCN7003
```

Draw a stacked bar plot for Region Sales group by “Genre” attribute
