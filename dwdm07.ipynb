Generate 100 random numbers for the age attribute and plot the equal-width and 
equal frequency histograms. Develop user defined functions to perform sampling of the 
age attribute: SRSWOR, SRSWR, and stratified sampling. Use samples of size 5 and the
strata “youth,”
“middle-aged,” and 
“senior.”

```python
import random
import matplotlib.pyplot as plt 
age=[]
for i in range (1,101): 
  age.append(random.randint(1,100))
age.sort()
n, bins, patches = plt.hist(age, edgecolor='black',bins=20) 
plt.xlabel('age')
plt.ylabel('Frequency') 
plt.show()
#19BCN7003
```

```python
# equal frequency
def equifreq(arr1, m): 
  a = len(arr1)
  n = int(a / m) 
  l=[]
  for i in range(0, m): 
    arr = []
    for j in range(i * n, (i + 1) * n): 
      if j >= a:
        break
      arr = arr + [arr1[j]] 
    l.append(arr)  #print(arr)
  return l

k=equifreq(age,20) 
print(k) 
c=['green']*20
plt.hist(k,edgecolor='black',bins=1,color=c) 
plt.xlabel('age')
plt.ylabel('Frequency') 
plt.show()
#19BCN7003
```

```python
import random
c = list(range(0, len(age))) 
def samSRSWOR(age):
  return random.sample(c,5) 
print(samSRSWOR(5))
#19BCN7003
```

```python
#Simple Random Sampling With Replacement import random
def SRWR(age):
  return random.choices( age, k = 14) 
print(SRWR(age))
#19BCN7003
```

```python
import pandas as pd
df = pd.DataFrame(age, columns =[ 'age'])
classify = []
for row in df['age']:
  if row <=25: 
    classify.append('youth')
  elif row <=45: 
    classify.append('middle-aged') 
  else: 
    classify.append('senior') 
df['classify']=classify
print(df)
#19BCN7003
```
