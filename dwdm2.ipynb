```python
import numpy as np
import time
start=time.time()
arr1=np.random.randint(10,size=(100,100))
arr2=np.random.randint(10,size=(100,100))
print("Array1\n",arr1)
print("\n")
print("Array2\n",arr2)
res= np.zeros((100,100),dtype=int)
for i in range(len(arr1)):
    for j in range(len(arr2[i])):
        for k in range(len(arr2)):
            res[i][j]+=arr1[i][k]* arr2[k][j]
print("Result\n",res)
end=time.time()
print("\n")
print("Execution time is: ",end-start)
#19BCN7003
```
